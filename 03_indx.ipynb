An Index in pandas is a label used to uniquely identify each row in a DataFrame. It can be a number, a string, a date, or any hashable type. It acts like a row identifier and helps with data alignment, selection, and filtering.

Now there are many ways to make a particular column as an index in pandas.

1. using index_col = ""
2. pd.set_index('Country',inplace=True)
3. Manually 
    df.index = df['Country']
4. Create a data frame with index
    data = {
    'Age': [25, 30],
    'Country': ['India', 'USA']
        }
    df = pd.DataFrame(data, index=data['Country']) 
5. .reset_index().set_index()
6. Multi indexing using 
    set_index(['Country', 'Year'],inplace=True)

```python
df = pd.read_csv("world_population.csv", index_col = "Country")
df
```

using index_col = "Country" we created Country as an index 
now if we need to reset it then we can use reset_index()

```python
df.reset_index(inplace=True)
df
```

Using set_index()

Now if we don't provide inplace= True here it will not save the changes we made so to keep the changes we made , we need to use inplace=True

```python
df.set_index("Country",inplace=True)
df
```

Lets search and filter using index - 
1. loc[] - non integer values 
2. iloc[] - integer values 

```python
df.loc["India"]
# df.iloc[1]
```

```python
df
df.reset_index(inplace=True)
df
```
