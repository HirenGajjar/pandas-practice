Filtering using Contains function
Q. Let say we want to find out all the countries that contains the "United" in Country column

Again - if we just use df['Country'].str.contains() then it will return boolean values 
So we need to use df[df['Country].str.contains('United')]

⚠ Also we use .str before .contains() because - df['Country'] is a series and .contains() is a string method. So to apply the string method on series we need to convert it into string first.

```python
# df['Country'].str.contains('United')

df[df['Country'].str.contains('United')]

```

Filtering based on Index -
There are two main methods 1.filter and 2. loc() and iloc()

Here we just made copy of the data frame to df2 where the index is based on Country column

```python
df2 = df.set_index('Country')
df2
```

We use filter method to filter with Country column, 
Filter takes an input of items=[] , and the values specified in the items will be filtered
Filter also takes regex
Filter takes like="" Values where it finds that pattern
Filter has an parameter called axis

axis=0 → filter rows 
axis=1 → filter columns (default)

```python
df2.filter(items = ['Continent','CCA3'],axis = 1)
```

Now using axis = 0 and search for specific country 

```python
df2.filter(items = ['India','Israel'], axis = 0)
```

```python
# Here we filtered through rows - using like ="di" string as we are using axis 0 and Country as an index

df2.filter(like="di", axis=0)

#Now we changed the axis to 1 and it will give all the column names that contains e in its name at any position
df2.filter(like="e",axis=1)
```

```python
df2.filter(regex="e$" ,axis = 1)

# Above regex "e$" will give the column titles and columns that ends with e - Growth Rate etc.
```

```python
df2.filter(regex='e$',axis=0)

# Where as having axis = 0 will filter through rows in country (as we have created it as an index in df2) and will give back all the countries that ends with 'e'
```

Now lets do Order by in original data frame 

```python
df
```

First we will get the data frame for rank upto 10 to make it easier to view

```python
df[df['Rank'] < 10]

#And then we will use sort_values()
# sort_values takes a parameter by="" and we did with Rank Column

df[df['Rank'] < 10].sort_values(by="Rank")
df[df['Rank'] < 10].sort_values(by="Continent")

#We can sort ascending or descending using another parameter called ascending = True or False

df[df['Rank'] < 10].sort_values(by = "Rank",ascending=False)


# we can sort using multiple columns and to do that we can make a list for by = []

# The following is multi level sorting - first it will sort based on CCA3 - then if two countries have same value for CCA3 then it will sort for Rank - if two countries have same rank then it will go for Continent
# Changing the order in list will change the output

df[df["Rank"] < 10].sort_values(by=["CCA3","Rank","Continent"] , ascending=False)

# We can also provide the ascending order for each sort condition in by = [] using ascending = [True, False,...]

df[df['Rank'] < 10].sort_values(by=["Rank","CCA3"],ascending=[False,True])

# here we did rank sorting in descending and CCA3 wih ascending
```
