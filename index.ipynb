Lets import text file
There are many ways to read text file
1. use read_csv and give sep="\t"
2. Proper way is to use read_table

```python
# df_text = pd.read_csv('countries_of_the_world_text.txt',sep='\t')
# df_text

df_text = pd.read_table('countries_of_the_world_text.txt')
df_text
```

We can also read the .csv file using read_table() with sep = ',' but it is better to use the method read_csv()

```python
df=pd.read_table('countries_of_the_world.csv',sep=",")
df
```
