```python
df.groupby("Liked").sum().sort_values(by="Flavor Rating",ascending=False)
```

```python
df.groupby("Base Flavor").min()
df.groupby("Base Flavor").max()
df.groupby("Base Flavor").sum()
df.groupby("Base Flavor").count()
```

We can do all the aggregation operation on group using agg() which takes the dictionary as an input


agg({
    "Column name" : ["List","Of",Operations"]
})


Not only for one specific column , but for multiple

```python
df.groupby("Base Flavor").agg({
    "Flavor Rating" : ['mean','max','count','min','sum'],
    "Liked" : ['max','count','min','sum']
})
```

```python
df.agg({
    "Total Rating" :['min','max','sum','std','median','mean','var','count']
})
```

Now lets do group by multiple columns   

```python
df.groupby(["Base Flavor","Liked"]).agg({
"Flavor Rating" :['mean','max','min','sum','std','var','count','median'],
"Texture Rating" :['mean','max','min','sum','std','var','count','median'],
"Total Rating" :['mean','max','min','sum','std','var','count','median']

})
```
